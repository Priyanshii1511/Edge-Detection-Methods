```python
import numpy as np
from scipy import ndimage
import cv2
from matplotlib import pyplot as plt
roberts_cross_v = np.array( [[ 0, 0, 0 ], [ 0, 1, 0 ], [ 0, 0,-1 ]] )
roberts_cross_h = np.array( [[ 0, 0, 0 ],[ 0, 0, 1 ], [ 0,-1, 0 ]] )
img=cv2.imread("image.jpg",0).astype('float64')
img/=255.0
verticle=ndimage.convolve(img,roberts_cross_v)
horizontal=ndimage.convolve(img,roberts_cross_h)
plt.imshow(img)
```

```python
edged_img=np.sqrt(np.square(horizontal)+np.square(verticle))
edged_img*=255
plt.imshow(edged_img)
```
