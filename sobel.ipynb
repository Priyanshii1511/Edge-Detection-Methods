```python
import cv2
from matplotlib import pyplot as plt
img=cv2.imread("image.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#original image
plt.imshow(img)
```

```python
#converting to grayscale
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
#blur the img for better edge detection
blur_img=cv2.GaussianBlur(gray_img,(3,3),1000)
plt.imshow(blur_img)
```

```python
#sobel egde detection it is performed for every axis
sobelx=cv2.Sobel(src=blur_img,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=31) #on x axis
plt.imshow(sobelx)
```

```python
sobely=cv2.Sobel(src=blur_img,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=31) #on y axis
plt.imshow(sobely)
```

```python
sobelxy=cv2.Sobel(src=blur_img,ddepth=cv2.CV_64F,dx=1,dy=1,ksize=31) #on x axis and y axis
plt.imshow(sobelxy)
```
