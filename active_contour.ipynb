```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("image.jpg")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
#blurring the image
blur_img = cv2.GaussianBlur(gray_img, (3, 3), 0)
plt.imshow(blur_img)
```

```python
#finding edges
edged_img = cv2.Canny(blur_img, 200, 100)
plt.imshow(edged_img)
```

```python
#finding the contour
contours,_=cv2.findContours(edged_img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
#creating copy of image for contour drawing
img_copy=img.copy()
#drawing the contour
pq=cv2.drawContours(img_copy,contours,-1,(0,255,0),2)
plt.imshow(pq)
```
